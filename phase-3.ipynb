```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data structure : [restaurant_id,2021,2022,2023,2024]
sales_data = np.array([
    [101, 150000, 160000, 170000, 180000],
    [102, 200000, 210000, 220000, 230000],
    [103, 250000, 240000, 230000, 220000],
    [104, 300000, 310000, 320000, 330000],
    [105, 180000, 190000, 200000, 210000]
])
print("==== Zomato sales analysis =====")
```

```python
print("sales data shape",sales_data.shape)
```

```python
print("sample data for 1st 3 restaurant:\n",sales_data[0:3,1:])
```

### total sales per year

```python
print(np.sum(sales_data[:,1:],axis=0))
```

### sales per restaurant

```python
# minimum sales per restaurant
print(np.min(sales_data[:,1:],axis=1))
```

```python
# maximum sales per year
print(np.max(sales_data[:,1:],axis=0))
```

```python
# average sales per restaurant
avg_sales = np.mean(sales_data[:,1:],axis=1)
print(avg_sales)
```

```python
#cumsum sales per restaurant
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
```

```python
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("average cumulative sales accross all restaurant")
plt.xlabel("years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
