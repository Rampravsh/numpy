```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1 = np.array([[1,2,3],[4,5,6]])
array2 = np.random.rand(3,3)
array3= np.zeros((4,4))

```

```python
np.save("array1.npy",array1)
np.save("array2.npy",array2)
np.save("array3.npy",array3)
```

```python
loaded_array1=np.load("array3.npy")
print(loaded_array1)
```

```python
try :
    logo = np.load("array2.npy")
    # display

    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("array")
    plt.grid("off")
except FileNotFoundError:
    print("numpy logo file not found")
```

```python
from PIL import Image
try:
    img = Image.open("dataset-thumbnail.png")
    img_array = np.array(img)
    np.save("numpy.npy",img_array)
    # print(f"successfully converted {img_array} to {"numpy.npy"}")
    print(f"array shape:{img_array.shape}")
    print(f"Array dtype:{img_array.dtype}")
except FileNotFoundError:
    print("file not found")
except Exception as e:
    print(e)
```
